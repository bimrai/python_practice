# Displaying Info

So we can dsplay info in the terminal via print. Our project is a tic tac toe game, so we will need a grid like UI. We can do this by stacking print on top of each other.

```python
print([1, 2, 3])
print([4, 5, 6])
print([7, 8, 9])
```

However we don't want this like that. We want it more modular. We can do this in a function.

```python
def display(row1, row2, row3):
    print(row1)
    print(row2)
    print(row3)
```

```python
example_row = [1, 2, 3]

display(example_row, example_row, example_row)
```

```python
# We can also do this

row1 = [' ', ' ', ' ']
row2 = [' ', ' ', ' ']
row3 = [' ', ' ', ' ']
```

```python
display(row1, row2, row3)
```

```python
row2[1] = 'X'
```

```python
display(row1, row2, row3)
```
