# Timing Python Code

```python
# we got two functions one and two that are the same but done differently
```

```python
def func_one(n):
    return [str(num) for num in range(n)]
```

```python
func_one(10)
```

```python
def func_two(n):
    return list(map(str, range(n)))
```

```python
func_two(10)
```

```python
# which one is more efficient?

# we can use the time method by importing it!
```

```python
import time
```

```python
# its in easy 3 steps

# function one: func_one() test

# CURRENT TIME BEFORE CODE
start_time = time.time() # pulls time from operating system
# RUN THE CODE
result = func_one(1000000) # million
# CURRENT TIME AFTER RUNNING CODE
end_time = time.time()
# ELAPSED TIME
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
# function two: func_two() test

# CURRENT TIME BEFORE CODE
start_time = time.time() # pulls time from operating system
# RUN THE CODE
result = func_two(1000000) # million
# CURRENT TIME AFTER RUNNING CODE
end_time = time.time()
# ELAPSED TIME
elapsed_time = end_time - start_time

print(elapsed_time)
```

```python
# timeit module

import timeit
```

```python
# you have to pass it as a string weirdly enough

stmt = '''
func_one(100)
'''
```

```python
setup = '''
def func_one(n):
    return [str(num) for num in range(n)]
'''
```

```python
timeit.timeit(stmt, setup, number=1000000)
```

```python
stmt2 = '''
func_two(100)
'''
```

```python
setup2 = '''
def func_two(n):
    return list(map(str, range(n)))
'''
```

```python
timeit.timeit(stmt2, setup2, number=1000000)
```

```python
# the %%timeit method only works on jupyter notebook, here is the syntax:

%%timeit
func_one(100)
```

```python
%%timeit
func_two(100)
```

```python
# throughout these tests function two is definitely faster
```
