## Simple Date Time Arithmetic

```python
# simple log in track info

# DATE

from datetime import date
```

```python
date1 = date(2026, 3, 2)
date2 = date(2025, 3, 2)
```

```python
# you can find out how much time has gone between say date1 and date2

time_gone = date1 - date2
```

```python
time_gone
```

```python
print(time_gone)
```

```python
print(time_gone)
```
