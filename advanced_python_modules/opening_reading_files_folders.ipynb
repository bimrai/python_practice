# Moving Files Around

```python
import shutil # short for shell utility module
```

```python
# say we want to move the practice txt file we created earlier into a folder. 

shutil.move('practice.txt', 'C:\\Users\\birah\\Desktop\\Files\\python_practice') # we pass the shutil.move(src, dst) dst for destination
```

```python
# now when we check the desktop which was where practice.txt was, its not there anymore. When we check
# the folder we placed it in, it has been moved to that folder.

os.listdir('C:\\Users\\birah\\Desktop\\Files\\python_practice')
```

```python
# we can check if practice.txt file is in our desktop or not to make sure its not duplicated but actually moved

os.getcwd()
```

```python
os.listdir()
```

As you can see, there is no longer any sign of practice.txt file, where as previously we had created it on the desktop with some strings.

If you hit os. + TAB, it lists a huge list of things you can do with this module.

There are 3 ways to deleting a file or files. 

1. os.unlink(path), deletes a file at the path you provide
2. os.rmdir(path), deletes a folder (folder must be empty) at the path you provide
3. shutil.rmtree(path), most dangerous, as it will remove all files and folders contained in the path so BECAREFUL USING THIS!

Fair warning, all of these methods cannot be reversed. Which means if you make a mistake, you won't be able to recover the file. Instead we will use the send2trash module. A safer alternative that sends deleted files to the trash bin instead of permanently removing the file/folders for practice.

Recommended: Install send2trash via pip install send2trash, where this will work with your bin over removing your files FOREVER...!

```python
import send2trash
```

```python
os.listdir()
```

```python
# lets bring back the practice file to the current directory

shutil.move('C:\\Users\\birah\\Desktop\\Files\\python_practice\\practice.txt', os.getcwd())
```

```python
# lets check again by running the same command os.listdir

os.listdir()
```

```python
# we can see the practice.txt file now.

# now we will send this to the trash or bin

# so we just use send2trash.send2trash( filename )

send2trash.send2trash('practice.txt') 
```

```python
# it isnt permanently gone, but has been put into the bin on your desktop.
# now when we list our directory, the file is not existing there anymore due to it being moved to the trash

os.listdir()
```

# Possibly most useful: os.walk

```python
os.getcwd()
```

```python
file_path = 'C:\\Users\\birah\\Desktop\\Example_Top_Level'
```

```python
# tuple unpacking

for folder, sub_folders, files in os.walk(file_path):
    print(f'Currently looking at {folder}')
    print('\n')
    print(f'The subfolders are: ')
    for sub_fold in sub_folders:
        print(f'\t Subfolder: {sub_fold}')

    print('\n')
    print('The files are: ')
    for f in files:
        print(f'\t File: {f}')
    print('\n')
```
