```python

# Character	Description	Example Pattern Code	Exammple Match
# \d	A digit	file_\d\d	file_25
# \w	Alphanumeric	\w-\w\w\w	A-b_1
# \s	White space	a\sb\sc	a b c
# \D	A non digit	\D\D\D	ABC
# \W	Non-alphanumeric	\W\W\W\W\W	*-+=)
# \S	Non-whitespace	\S\S\S\S	Yoyo
```

```python
text = "My phone number is 408-555-1234"
```

```python
phone = re.search('408-555-1234', text) # normal way
```

```python
phone
```

```python
# pattern method

phone = re.search(r'\d\d\d-\d\d\d-\d\d\d\d', text) # should match the same number 408-555-1234
```

```python
phone
```

```python
# try with another number to test

text = "My phone number is 117-343-2007"
phone = re.search(r'\d\d\d-\d\d\d-\d\d\d\d', text)
phone
```

```python
# we can return all matched again by using group

phone.group()
```

```python
# we can use quantifiers to indicate repetitive or mundane repititions 

# Character	Description	Example Pattern Code	Exammple Match
# +	Occurs one or more times	Version \w-\w+	Version A-b1_1
# {3}	Occurs exactly 3 times	\D{3}	abc
# {2,4}	Occurs 2 to 4 times	\d{2,4}	123
# {3,}	Occurs 3 or more	\w{3,}	anycharacters
# \*	Occurs zero or more times	A\*B\*C*	AAACC
# ?	Once or none	plurals?	plural
```

```python
phone = re.search(r'\d{3}-\d{3}-\d{4}', text)
```

```python
phone
```

```python
phone_pattern = re.compile(r'(\d{3})-(\d{3})-(\d{4})')
```

```python
results = re.search(phone_pattern, text)
```

```python
results
```

```python
results.group()
```

```python
re.search(r"cat", "The cat is here")
```

```python
re.search(r"cat", "The dog is here") # no match
```

```python
# what if we wanted cat or dog? we use the OR operator

# however first, lets look at our output. its showing at multiple times due to multiple words ending in 'at'.
re.findall(r'at', 'The cat in the hat sat there.')
```

```python
# what if we want cat and not multiple words that could end in 'at'?
# we can use the wildcard '.' before 'r'.

# the . is a wildcard where we fetch any specific letters where it expects one before in this case
re.findall(r'.at', 'The cat in the hat sat there.')
```

```python
# say we have instead of 3, but a longer word like splat?

# this will only return the at and a singular letter before 'at'.
re.findall(r'.at', 'The cat in the hat went splat!')
```

```python
re.findall(r'..at', 'The cat in the hat went splat!') # add one more wildcard it will return the second letter of splat
```

```python
re.findall(r'...at', 'The cat in the hat went splat!')
# add another and we finally get the full word for splat, however we are also getting the e space or white space from the words before
# like 'the' and again 'the' both before cat and hat. 
```

```python
re.findall(r'^\d', '1 is a number') # indicates its searching for a number in the start of the string, so ^ means STARTS WITH NUMBER
```

```python
re.findall(r'^\d', 'one is a number') # returns nothing
```

```python
re.findall(r'\d$', 'The number is 2') # the dollar sign at the end means ENDS WITH NUMBER
```

```python
phrase = 'There are 3 numbers 34 inside 5 this sentence'
```

```python
pattern = r'[^\d]'
```

```python
re.findall(pattern, phrase)
```

```python
pattern = r'[^\d]+'
# the + returns it back to normal in list like object
```

```python
re.findall(pattern, phrase)
```

```python
test_phrase = 'This is a string! But it has punctuation. How can we remove it?'
```

```python
re.findall(r'[^!.?]+', test_phrase)
```

```python
clean = re.findall(r'[^!.?]+', test_phrase)
```

```python
' '.join(clean) # makes it a string again
```

```python
text = 'only find the hyphen-word in this sentence. But you do not know how long-ish they are'
```

```python
# find words with hyphen in them

pattern = r'[\w]+'
```

```python
re.findall(pattern, text)
```

```python
pattern = r'[\w]+-[\w]'
```

```python
re.findall(pattern, text)
```

```python
# removing the square brackets gets back the same results

pattern = r'\w+-\w'
```

```python
re.findall(pattern, text)
```
