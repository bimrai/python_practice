# Zipping and Unzipping Files

```python
f = open('fileone.txt', 'w')
f.write('One File')
f.close()
```

```python
f = open('filetwo.txt', 'w')
f.write('Two File')
f.close()
```

```python
import zipfile
```

```python
# compress file

compress_file = zipfile.ZipFile('compressed_files.zip', 'w')
```

```python
compress_file.write('fileone.txt', compress_type=zipfile.ZIP_DEFLATED) # this has now been compressed and zipped into the zip file 
```

```python
compress_file.write('filetwo.txt', compress_type=zipfile.ZIP_DEFLATED) # second file is now with fileone in the zip file/folder
```

```python
compress_file.close() # closes the zipfile
```

```python
# extract 

zip_obj = zipfile.ZipFile('compressed_files.zip', 'r')
```

```python
# if you want to extract all
zip_obj.extractall('extracted_content') # we pass in a string which is the name of the folder it will be extracted to
```
