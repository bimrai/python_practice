## Control Flow using If, Elif, Else

If condition is if this then that, that being the return. Elif is else if, else is the final output to handle where neither if or elif statemets are met.

```python
# if
# elif
# else

# Python contorl flow syntax utilises indentations (whitespaces) and colons
# This indentation system is crucial to Python and is what sets it apart from other programming languages
```

```python
# if example

#if some_condition_here:
    # then execute code if condition true

if 1 == 1:
    print('This condition is true so the if statement runs.')
```

```python
# else example

# if some_condition_here:
     # then execute code if condition true, if false then go to else
#else:
     # do something else

if 1 == 2:
    print('This condition is false so moves to else.')
else:
    print('If was not met, so we move onto else.')
```

```python
# elif example

# if some_condition_here:
     # then execute code if condition true, if false then go to else
# elif some_other_condition:
     # execute something different
# else:
     # do something else


if 1 == 2:
    print('This condition is false so moves to else.')
elif 1 == 1:
    print('Elif condition was met so we print this comment')
else:
    print('If and Elif was not met, so we move onto else.')
```

```python
hero = True
power = True

if hero and power:
    print('A true hero')
```

```python
hero = True
power = False

if hero and power:
    print('A true hero')
else:
    print('Not a hero yet. Keep training!')
```

```python
hero = True
power = False

if hero and power:
    print('A true hero')
elif hero and not power:
    print('Hero spirit is strong, must work on powers!')
else:
    print('Not a hero yet. Keep training!')
```

```python
# You can stack elifs together too

loc = 'Game'

if loc == 'Auto Shop':
    print('Cars are cool')
elif loc == 'Bank':
    print('Got to take out some cash!')
elif loc == 'Game':
    print('I want to buy a game!')
else:
    print('I will stay home.')
```
