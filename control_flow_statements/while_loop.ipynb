## While Loop

The while loops will keep running the code while some conditions True.

```python
# Syntax

#while:

# while some_boolean_condition:
#     do something


# while else:

# while some_boolean_condition:
#     do something
# else:
#     do soemthing different
```

```python
# Example 1

x = 0

print('While x is smaller than 5, keep doing x = x + 1 and printing it out every iteration')
print('-----------------------')

while x < 5:
    x += 1 # same as x = x + 1
    print(f'Current value now is {x}')
```

```python
# You can also add else as mentioned above
x = 0

while x < 5:
    # if we print this first we see 0 in the output, if we print after x += 1 then we just see 1 as +=1 was performed first 
    print(f'Current value of x is: {x}.')
    x += 1 # same as x = x + 1
else:
    print(f'The variable x \"{x}\" is not less than 5.')
```

```python
# Same code, lets check if instead of x, we use y and it's above 5 what happens

y = 10 # it is now 10 but the condition is less than 5


while y < 5: 
    print(f'Current value of x is: {y}.')
    y += 1
else:
    print(f'The variable x \"{y}\" is not less than 5.')
```

## break, continue, pass

Break, continue and pass statements add additional functionality for various cases. The three statements are defined by:

break: Breaks out of the current closest enclosing loop.

continue: Goes to the top of the closest encolsing loop. Other words,  continue essentially means skip this block and continue to finish loop.

pass: Does absolutely nothing at all. Does not even throw an error. 

## Pass:

```python
x = [1, 2, 3, 4, 5]

for items in x:
    # comment <- leaving comment here and trying to move on/output with while loop will throw error: _IncompleteInputError: incomplete input
                 # this is because the while loop is expecting something to happen but a comment is not a function. Hence why error is thrown
    pass # however this moves on with the code without errors thrown

print('Python script reaches end of code.') # to test that it moves on and still runs without breaking
```

## Continue:

```python
mystring = 'Peter'

for letter in mystring:
    if letter == 't':
        continue # continue essentially says skip this block and continue to finish loop
    print(letter)
```

## Break:

```python
name = 'John'

for letter in name:
    if letter == 'h':
        break # break essentially stops or breaks the loop / block of code. Here it is under the condition if h is met when iterated, break
    print(letter)

# as you can see after condition was met, 'h', we stopped at h and output Jo
```

```python
# Break with a while statement as its useful for it

z = 0

while z < 5:

    if z == 2: # sets if condition, when we meet 2 what shold it do
        break  # within if condition, it tells us to break so it will stop on 2 but not include 2 & it wouldnt meet print(z). Stopped before
    
    print(z)
    z += 1
```
