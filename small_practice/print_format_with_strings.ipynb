When using format() method, you use {} as a placeholder for the string within the .format() method. Here we placed {} in the initial string, then after the string ends we add the .format() with the value or argument 'STRING'. This then replaces the curly brackets with the string within the format() like below:

```python
print('This is a {} type!'.format('STRING'))
```

You can also insert strings by the index position. You insert the index position of format() within the curly braces to order which word should sit within the string. Below is an example:

```python
print("I'm your {1} {2} {0}!".format("Spider-Man", "friendly", "neighbourhood"))
```

```python
# You can repeat the same word within format by repeating the index value within the curly braces like so

print("{0}, {0} does whatever a {1} can!".format("Spider-Man", "spider"))
```

You can also assign key words, similar to assigning variables like so:

```python
print("I'm your {b} {c} {a}!".format(a = "Spider-Man", b = "friendly", c = "neighbourhood"))
```

Moving on to float formatting which follows "{value:width.precision f}

```python
result = 100/777
result
```

```python
# We can change the level of precision we want change the width of the number itself. Check it out here below!

print("The result was: {}".format(result))
```

```python
# You can also do it like so:

print("The result was: {r}".format(r = result))
```

Referring back to the {value:width.precision f}, value is value itself, for example we will use r that holds 100/777. The width is the spacing or white space between characters or numbers and the precision value. Below is the example.

```python
print("The result was: {r:1.3f}".format(r = result))

# As you can see the output is 0.129. This means the precision starts from after decimal place. So after decimal is 1 which is 0 in the index
# 2 is 1 in index, 9 is 2 in the index and 3 is 7 in the index. 7 is the stop but not incuded hence it cuts off 7 and it also rounds up the 8 to 9 due to 7 being
# closer to the significant number making 8 to 9.
```

```python
# Just to give example of the width in action if we want to add space
print("The result was: {r:10.3f}".format(r = result))
# As you can see the width is just before the value within the curly braces
```
