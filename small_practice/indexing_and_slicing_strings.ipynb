```python
# In indexing, first character, even if it is white space will count as 0. Index always starts from 0.

# if we say:
# 'Hello World'
#  012345678910

# Here H is 0, e, is 1 and under we get to o being 4, after o is white space but it will be 5. Only then when we move on to W it is 6. So the index numbering convention is strict even if whitespaces exist between characters.
```

```python
my_string = 'Hello World'
```

```python
my_string
```

If you want to grab the first letter, remember it starts with 0 and in our case with the variable string above, H is the first letter and there is no white space before it.

In order to return the value using indexing, in this case we created a variable above already named my_string, holding the value 'Hello World'. So the first step is to type out the variable name we want to pull the value from. Then use [] after the variable like so -> my_string[]. After that, you just input the index number to pull a specific value out from the string via indexing.

```python
my_string[0]
```

Again as explained in the start, if we want to return the character 'W', we use the index number of 6

```python
my_string[6]
```

```python
# We can also use negative indexing where the initial character or white space remains 0, however it is in reverse from then.
# Here is an example.
# So we have say a string:
# 'I fly'
#  0,-4,-3-2,-1
# This may look confusing but essentially, we are working in reverse so from the last character in the string, where -1 is y and
# anything going backwards on string will keep going in negatives, -1, -2, -3. So l would be -2, f would be -3, the white space -4
# but the first character I remains to be 0.
```

So say we want to get 'd' from the my_string using reverse indexing, we would use -1 because in 'Hello World', d is the last character but in reverse indexing it would be the negative first, aka -1. Check it out.

```python
my_string
```

```python
my_string[-1]
```

```python
# So if we want H
my_string[0]
```

```python
# and if we want the character 'W'
my_string[-5]
```

```python
# So notice how if we use 1 and -1 in this case just to gain better clarity. This is a basic example of the differences between
# an index and a reverse index.

# Remember we are using 'Hello World' from the variable my_string
my_string
```

```python
# Normal index
my_string[1] # this will return 'e', as in 'Hello', 'H' is 0, 'e' is 1.
```

```python
# Reverse index
my_string[-1] # this will return 'd' from the very end of the string from 'Hello World'.
```

```python
# IMPORTANT - no matter what, index or reverse index, the index value of 0 will always remain the first character in the beginning.
#             in our case for the string 'Hello Word', this will always be the value will always be 'H' for the index 0
my_string[0]
```

Let's work with slicing now.

```python
string = 'abcdefghijk'
```

```python
# Say we want to start from 'c'. We still use the index method here. So we start from 2.
string[2]
```

```python
# Say we want to get from 'c' and everything after it. We use the colon symbol :
string[2:]
```

```python
# So now we want to limit how much we grab after colon since we don't always want to grab everything else to make it cleaner.
# We must add another index value after colon. But here is the catch when slicing in this step.
# The first value or known as start value is easy because we pick the index start value. That index start value includes the value.
# However when we choose an ending value, it excludes the value. So in this example we want 'abc' only.
# So the starting index value would be 0 for the string 'a' but the ending index value would be 3 which would be 'd'.
# We choose d because it will stop at 'd', but exclude 'd'. It will not keep it within the slicing values. 
# So essentially, slice is [start:stop/exclude value]. So in abcdefghijk, we choose 'a' for our start and 'd' for our stop so we
# cut off d and keep everything before. 
string[0:3]
```

```python
# After undersanding the fundamentals, you can also short cut it like this:
string[:3]
# By not providing start value, python automatically assumes 0 and you can find a stop/exclude point in our case 'd' to output 'abc'
# Best way to reiterate - slice[start:stop up to but not including]
```

```python
# One more practice, say we want 'def' only from the string variable 'abcdefghijk', we figure out the start. 'd' is 3
# and remember we don't use the index value for 'f' which would be 5 as that would not include 'f'. So we use 6 which is 'g' after 'f'
# which would exclude or not include 'g'. Hence keeping 'def'. Let's test it.

string[3:6]
```

```python
# Okay very last practice using this method. Let's try get 'jk'
string[9:11] # here we use 11 even if the character doesn't exist. That way we include 10th index value.string
```

```python
# To include all the strings itself
string[::]
```

```python
# You can do whats known as 'STEP SIZE' where it will skip to certain characters within a string when slicing like so
string[::2]
# This will start from 0 and move every 2 values and include that. So 0 is 'a', we jump to index 1 of 'b' which is 1
# we then jump to index 2 which is 'c' but includes that. It will go on until the end.
```

```python
# A more complexed example
string[1:6:2]
# string[start:stop but not include:step size] So we start from index 1, stop and exclude index 6,
# jump every 2 value including that landing value
```

```python
# You can reverse your string this way too
string[::-1]
# This is saying from the beginning to the end, take a single step backwards since step back is only -1. Remember, negative is reverse index
```
