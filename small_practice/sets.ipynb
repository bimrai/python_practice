# Sets
Sets are unordered collections of unique elements. Just like in discrete mathematics, when you have a set A = {a, b, c} and B = {b, c, d, e, f, g}, the U (union) of A and B (A U B) = {a, b, c, d, e, f, g}. There can be no duplicates in sets for Python, hence the unique elements. 

```python
my_set = set()
my_set
```

```python
my_set.add(1)
my_set
```

Bear in mind that the output above, it looks like a dictionary due to curly brackets being the reference but for it to be a dictionary it requires a key:value pair to meet the condition of being a dictionary. However, this is just an unordered collection of any items you want.

```python
# We can add another element or value again.
my_set.add(2)
my_set
```

```python
# However, notice if you add 2 AGAIN, it will not print two amounts of 2's. It will only show one of 2 element due to
# sets being unique, not allowing repitition of elements. SO, if you try add a value that's already in there, it will not repeat it
my_set.add(2)
my_set
```

```python
# Here we can check with list to see that sets only keep unique elements when returning or outputting

mylist = [1, 2, 1, 2, 1, 2] # We have a list of three 1's and three 2's. 
set(mylist) # We put it in the set method and now watch as the rest of the repeated elements vanish! 
```
