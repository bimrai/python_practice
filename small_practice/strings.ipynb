If we say input two different strings, we only get the last written string

```python
'Hello World'
'Hello World 2'
```

So if you want to print all strings out in this case, wrap within print.

```python
print('Hello World')
print('Hello World 2')
```

You can also create a new line, within inline or in sring using \ (backslash) and n like this: \n

```python
print('Hello \n World')
```

There is a slight space before World and due to our space after \n, it prints out the way it does above. To avoid that, python is white space sensitive so we can remove the space between \n and World like so below.

```python
print('Hello \nWorld')
```

There is also another feature using the \t for tab, equivalent to 4 spaces.

```python
print('Hello \tWorld')
```

The function len() returns the length of a string. A string say 'Hello' is 5. This is because there are 5 characters in Hello. Check it out below.

```python
len('Hello')
```

Now bear in mind that white spaces do count as characters in this case. Especially when referencing character index too. But for now let's check the white space (or just space in simple terms) whether its being recognised as a character when using len(). We will use the string 'I fly'. Here you only see 4 real characters, however there is white space between I and f. This space counts as a character. So if you create 2 spaces or 4, they will add on with the number of real characters used too. Check it out below.

```python
len('I fly')
```
