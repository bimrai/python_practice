```python
# mode = 'r' is read only
# mode = 'w' is write only (will overwrite files or create new)
# mode = 'a' is append only (will add on to files)
# mode = 'r+' is reading and writing 
# mode = 'w+' is writing and reading (Overwrites existing files or creates a new file!)
```

```python
%%writefile new_file.txt
ONE ON FIRST
TWO ON SECOND
THREE ON THIRD
```

```python
# read mode
with open('new_file.txt', mode='r') as file: # <- clarify this is the variable for file, then in function we use it. file = new_file.txt
    print(file.read())
```

```python
# add mode
with open('new_file.txt', mode='a') as file: # using append makes cursor all the way at end of text in file and make a new line when we add
    file.write('FOUR ON FOURTH')
```

```python
# Once change is made, call or return read function 

# read mode
with open('new_file.txt', mode='r') as file: # <- clarify this is the variable for file, then in function we use it. file = new_file.txt
    print(file.read())
```

```python
# writing mode r

with open('eifosifeif.txt', mode = 'w') as file:
    file.write('I CREATED THIS FILE!')
```

```python
with open('eifosifeif.txt', mode='r') as file:
    print(file.read())
```
