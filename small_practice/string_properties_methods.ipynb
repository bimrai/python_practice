```python
name = "Bim"
```

```python
name
```

```python
# Strings are immutable. The code below will throw an error.
name[0] = 'T'
```

```python
# We can concat as a method of mutating a string. Let's try make Bim to Tim
last_letters = name[1:] 
last_letters
```

```python
# We can now concat using +, 'T' + last_letters = 'Tim'. Let us just check it.
'T' + last_letters
```

```python
# We can store it in a variable too. 
new_name = 'T' + last_letters
new_name
```

```python
# We can concat strings together and also reassign the x to a new value.
x = 'Hello World.'
x + ' It is so beautiful!'
```

```python
#reassign
x = x + ' It is so beautiful!'
x
```

```python
# We can also use multiply symbol * like so for strings. Let's say a bear is sleeping.
bear = 'z'
bear * 3
```

```python
# Then again, store it in a variable
lion = 'z'
lion = lion * 3
lion
```

```python
# At this stage, obvious but string and integers don't calculate on this level. However, we can concat strings together like so:
'2' + '4' # This will make a string '24' not 2 + 4 = 6. 
```

```python
# There are multiple methods built in python. First define to utilise these methods. 
sentence = 'hello world'
# sentence variable has all lower case. We can use one of many methods to for example make them all upper case.
sentence.upper()
```

```python
# If you do not want to change the original string you should always define or reassign
# Also don't forget when using or calling methods, the parenthesis () should be after the method. upper().
game = 'minecraft'
capital_game = game.upper()
print(f'Original variable: {game}')
print(f'Reassigned variable with upper method: {capital_game}')
```

```python
# We can also lower case
capital = 'UNITED KINGDOM'
capital.lower()
```

```python
# You can also split creating a list of the strings
list = 'Apple Banana Orange Mango'
list.split()
```

```python
# With the same split, we can exclude letters like so
cat = "Wio wio wio"
cat.split('o')
```
