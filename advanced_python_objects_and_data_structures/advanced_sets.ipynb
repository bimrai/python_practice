```python
# you can use copy

s = {1, 2, 3}
```

```python
# sc for set copy
sc = s.copy()
```

```python
# as you can see it doesnt change or effect the original set
print('s is: ', s)
print('sc is now: ', sc)
```

```python
s.add(4)
print(s)
print(sc)
```

```python
# differences between sets, we check s with another passed on set in this case sc, 4 is the only value not in both but only in sc.  
s.difference(sc)
```

```python
# intersection

s1 = {1, 2, 3}
```

```python
s2 = {1, 2, 4}
```

```python
s1.intersection(s2)
```

```python
# we can update s1 as the new set from s1 intersection s2, so s1 will hold 1 and 2 only

# current s1
s1
```

```python
# we can use the update version on the intersection too like so 
s1.intersection_update(s2)
```

```python
# now when we call s1, remember earlier intersection of s1 and s2 was {1, 2}, lets see what s1 is
s1
```

```python
# "is this joint" method, returns TRUE if the given 2 sets do not share elements, else false
s1 = {1, 2}
s2 = {1, 2, 4}
s3 = {5}
```

```python
s1.isdisjoint(s2)
```

```python
s1.isdisjoint(s3)
```

```python
# check subset, means where a sets elements are all within a bigger set
s1.issubset(s2)
```

```python
#super set, whether a set contains another set
s2.issuperset(s1)
# returns true since elements 1 and 2 exist in s1 and s2 is a bigger or super set
```

```python
# symmetric difference: opposite of intersection, any non shared values between sets
s1
```

```python
s2
```

```python
s1.symmetric_difference(s2) # should output only 4 from sets s1 and s2 
```

```python
# you can also change and update say s1 via symmetric difference
s1.symmetric_difference_update(s2)
s1 # is now just 4
```

```python
# union returns all sets in a and/or b
s1.union(s2)
```

```python
# we can update too for union
s1
```

```python
s1.update(s2)
```
