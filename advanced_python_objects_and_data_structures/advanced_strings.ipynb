# Advanced Strings

```python
s = 'hello world'
```

```python
# capitalises first word only within a string
s.capitalize() 
```

```python
# upper case all string
s.upper()
```

```python
# lower case all string
capital_s = 'HELLO WORLD'

capital_s.lower()
```

```python
# counts occurences in a string say 'o' in this example
s.count('o')
```

```python
# finds only the first appearance of say 'o' in this example via index
s.find('o')
```

```python
# string check 
s
```

```python
# it centers the string between say z in this example
s.center(20, 'z')
```

```python
# back slash \t create escape tab

print('hello\thi')
```

```python
# can also do expand tabs method
'hello\thi'.expandtabs()
```

```python
s = "hello"
```

```python
# checks alpha numeric
s.isalnum()
```

```python
s.isalpha()
```

```python
# is lower returns true if all case character is lower case

s.islower()
```

```python
# looks for space but we dont have any white spaces in s variable 
s.isspace()
```

```python
# only returns true if s is a title case string
s.istitle()
```

```python
s.isupper()
```

```python
'HELLO'.isupper()
```

```python
s.endswith('o') # ends with x letter so we check o with hello in the s variable
```

```python
# same as doing
s[-1] == 'o'
```

```python
# split

s
```

```python
s.split('e')  # returns everything before and after but not including the chosen letter to split from 
```

```python
# split splits every instance of the letter you pass it
string = 'hiihihhhhiihhhhhihihihihi'
string.split('i')
```

```python
string.partition('i')
# takes first letter, separates second letter from first and third, third section is the rest after the second partitioned letter
```
