## Chaining Comparison And Logical Operators

```python
# We can use logical operators to combine comparisons:
# and
# or
# not
```

```python
# We can chain like so
1 < 2 < 3
```

```python
# However all conditions must be true, for outut to be true.
# Here is how we check
1 < 2 > 3 # This will return false. Despite 1 < 2 first half being true, final half is 2 > 3 which is false
```

```python
# You can use a logical operator to combine comparison operators

# and means both side must be true for output to be true. 
# 1 < 2 is true but 2 > 3 is false, hence overall statement is false
1 < 2 and 2 > 3
```

```python
1 < 2 and 2 < 3 # true because both sides are true
```

```python
'string' == 'string' and 2 == 2 # mix with different data types
```

```python
('string' == 'string') and (2 == 2) # some people prefer to use parentheses for readability and some libraries use them too for this
```

```python
# or makes sure one side is at least true to return true, or both true to return true
# it only returns false when both comparisons on each sides are false

# both being true returns true
1 == 1 or 2 == 2
```

```python
1 == 1 or 'hello' == 'goodbye'
```

```python
# both sides are false here
2.5 == '2.5' or 'hello' == 1
```

```python
# not logical operator returns opposite boolean

# the two strings are both 'a' being compared which will return true
'a' == 'a'
```

```python
# However using not with it returns false despite being true.
not('a' == 'a')
```

```python
1 == 1 # is true
```

```python
# Now when we use not

not 1 == 1
```

```python
500 > 1000 # is false
```

```python
# But not in front of it returns true, the opposite
not 500 < 1000
```

```python
# not is similar to the ! comparison operator where it returns opposite
2 != 2
```
