# Web Scraping Libraries

```python
import requests
```

```python
import bs4
```

```python
result = requests.get("https://webscraper.io/test-sites/e-commerce/allinone") # from import requests
```

```python
type(result)
```

```python
result.text
```

```python
soup = bs4.BeautifulSoup(result.text, "lxml") # cleans up the messy data (from import b4s)
```

```python
soup
```

```python
soup.select('title')
```

```python
soup.select('p')
```

```python
# because theres a lot of p and it returns as a list, we can use the index select list or element to pick and choose what text we want

soup.select('p')[0]
```

```python
soup.select('p')[1]
```

```python
# as you can see above we picked the first 2 p tags, now the third is marketplace
soup.select('p')[2]
```

```python
# if you want say just the first p tag "<p>Cloud Scraper</p>" but as a string without it being an element within HTML do this:

soup.select('p')[0].getText()
```

```python
# you can assign tag elements too
site_paragraphs = soup.select("p")
```

```python
site_paragraphs[0]
```

```python
site_paragraphs[0].getText()
```
