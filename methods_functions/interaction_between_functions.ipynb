# Interaction Between Functions - Three Cup Monte

Three Cup Monte is just a cup shuffling game where you guess the ball once the ball is hidden and shuffled. In our case, we are just doing the scripting so there is no ball or cups to visually please our user for now hahahaha. It is just a python list with two empty strings, '', and a single string with 'o', to represent our ball.

```python
# how to shuffle a list in python. It is a built in function shufle()

example = [1, 2, 3, 4, 5, 6, 7]
```

```python
# importing the shuffle from random library 

from random import shuffle
```

```python
# pass example which the list we created above into shuffle

shuffle(example)
```

```python
ball = ['', 'O', '']
```

```python
def shuffle_list(ball):
    shuffle(ball)
    return ball
```

```python
result = shuffle_list(ball)
```

```python
# We now make a player guess so take in input

def player_guess():

    guess = ''

    while guess not in ['0', '1', '2']: # reason list of choice is string in while loop is because input returns string so must match
        guess = input('Pick a number: 0, 1, or 2')

    return int(guess) # we then convert that string to a integer afterwards to match the index place of shuffle_list
```

```python
# run test of the function - if unwanted input is entered, it will keep asking until right value is entered then it stops.

player_guess()
```

```python
# assign the guess value to a variable

chosen_index = player_guess()
```

```python
# check the variable to see if it has stored safely

chosen_index
```

```python
# now we combine the functions to check the player guess with the balls location

def check_guess(ball, guess):
    if ball[guess] == 'O':
        print("You guessed it right!")
    else:
        print("Unlucky. Try again...")
        print(ball)
```

```python
# now the logic 

# INITIAL LIST

ball = ['', 'O', '']

# SHUFFLE LIST

mix_ball = shuffle_list(ball)

# USER GUESS

guess = player_guess()

# CHECK GUESS AND LIST

check_guess(mix_ball, guess)
```
