# Basics of Python Functions

```python
def say_hello():
    print('Hello')

say_hello()
```

```python
def say_hello():
    print('Hello')
    print('What\'s up?')

say_hello()
```

```python
say_hello # without parenthesis it will state what it is, in this case its a function, below means you aren't calling the function
```

```python
def say_hello(name):
    print(f'Hello {name}')

say_hello('Bim') # if empty it will throw error and ask for a positional argument, a value essentially
```

```python
def say_hello(name = 'Bim'): # You can even add a value in function and call without argument later
    print(f'Hello {name}')

say_hello() # call without argument if function has variable value 
```

```python
def add_num(a, b):
    return a + b

add_num(2, 6)
```

### Key point: Using return over print allows you to assign it into a variable.

Now say we did, "return a + b", we can then assign the result of what we called into a variable. 

Here is a n example:

```python
def add_num(num1, num2):
    return num1 + num2

result = add_num(6, 4)
```

```python
# Then here we can just use the variable else where holding the result, in this case 10 is the value due to 6 + 4 = 10

result
```

```python
# print explanation to why function dont use print much

def print_result(a, b):
    print(a + b)
```

```python
# check it we can store print result call

result = print_result(10, 5)
```

```python
# call test

result # we dont get anything back
```

```python
# Here is why

type(result) # it is a none type
```

```python
# return example to why its used more often

def return_result(a, b):
    return a + b
```

```python
return_result(20, 30)
```

```python
result = return_result(20, 30)
```

```python
result
```

```python
# you can add print and return together but its not common at all

def add_function(a, b):
    print(a + b)
    return a + b

result = add_function(5, 1)
```

```python
# We can check what the data types are rather than just a and b in the argument.

# Python is dynamically typed, meaning you don't need to specify beforehand what data type to expect num1 and num2 to be.
# This allows you to program faster but exposes you to a lot of bugs.
# Similar to JS, fast in development, bugs are more prone. Hence Typescript was implemented.

# example

def sum_numbers(num1, num2):
    return num1 + num2
sum_numbers('10', '20') # this just concatenates the number as they're string. Not int. We are takin guser input, which is a string.
```

##### So when you are making software, you should always check the type of the data before returning the result.
