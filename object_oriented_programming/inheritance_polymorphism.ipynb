# Inheritence and Polymorphism

## Inheritence

```python
class Animal():

    def __init__(self):
        print("Animal Created")

    def who_am_i(self):
        print("I am an Animal")

    def eat(self):
        print("I am eating")
```

```python
# Make an instance of an animal
myanimal = Animal()
```

```python
# We add_who_am_i method
myanimal.eat()
```

```python
myanimal.who_am_i()
```

Newly formed classes can use the Animal class to inherit some of its method if they want to use it again.

```python
# example, and you think 'hey alot of the animal features are useful to the dog or are similar'
# we add the class name we want to derive features and methods from. It is now known as a DERIVED CLASS.

class Dog(Animal):

    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")

    # overwrite older methods
    def who_am_i(self):
        print("I am a Dog.")

    # can add additional methods related to Dog only despite being an animal
    def bark(self):
        print("WOOF!")

    # replace the eating with a more relavent print or info/data
    def eat(self):
        print("I am a dog and I am eating.")
```

```python
# it will output animal created and dog created. this shows that it was pulled from the base class printing the initialisation,
# then the dog class derived from the animal class to gain access to animal class methods etc, which it also runs the initialisation print
# message similar to animal created, but its in Dog class saying "Dog created"

my_dog = Dog()
```

```python
# So we do not have eat or who am i defined within the dog class itself, but because it derives these methods and such, dog class has gained 
# access to it. lets check it out

my_dog.eat()
```

```python
my_dog.who_am_i()
```

```python
# You can also overwrite older methods. previously it was pulled from the animal method, we overwrote the old method for who_am_i
# within the dog class itself and it will print "i am a dog" since we are calling from the dog class that is derived from animal class but
# a lot more updated

my_dog.who_am_i()
```

```python
my_dog.bark()
```

```python
my_dog.eat()
```

## Polymorphism

In python, polymorphism refers to the way in which different object classes can share the same method name, and then those methods can be called from the same place, even though a variety of different objects might be passed in.

```python
# example

class Dog():

    def __init__(self, name):
        self.name = name

    def speak(self): 
        return self.name + ": " + "Woof!"   
```

```python
class Cat():

    def __init__(self, name):
        self.name = name

    def speak(self):
        return self.name + ": " + "WI Wi wi!"  
```

```python
buddy = Dog("Buddy")
felix = Cat("Felix")
```

```python
print(buddy.speak())
```

```python
print(felix.speak())
```

### Both dog and cat class have the speak method. When called, each object to the speak method returns a result thats unique to the object. In other words, it is only unique to the dog to say woof and the cat to say wiwiwi, as well as their names too. There is a few different ways to demonstrate polymorphism. Like a for loop!

```python
# example of polymorphism, they both share the method speak, but they are different types.
# one will list main.Dog, the other will list main.Cat.

for pet in [buddy, felix]:

    print(type(pet))
    print(type(pet.speak()))
```

```python
for pet in [buddy, felix]:

    print(type(pet))
    print(pet.speak()) # when we remove the type away to run the method for both instances it will show for both where they belong
```

```python
# A more common way over the iteration way is by a function

def pet_speak(pet):
    print(pet.speak())
```

```python
pet_speak(buddy)
```

```python
pet_speak(felix)
```

### Abstract Class

An abstract class is one that never expects to be instantiated. One that never expects to create an instance of this class, instead designed to only serve as a base class

```python
# example again, never expected to be instantiated, only works or serves as a base class

class Animal():

    def __init__(self, name):
        self.name = name

    def speak(self):
        raise NotImplementedError("Subless must implement this abstract method.")
```

```python
myanimal = Animal("Po")
```

```python
myanimal.speak() # not expected to use Animal in that fashion
```

```python
# Instead is expected to:

# pass in animal in subclass
class Dog(Animal):

    def speak(self):
        return self.name + " says woof!"
```

```python
# same for the cat

class Cat(Animal):

    def speak(self):
        return self.name + " says meow!"
```

```python
lad = Dog("Lad")
luna = Cat("Luna")
```

```python
lad.speak()
```

```python
luna.speak()
```
