```python
class Dog():

    # class object attribute, this is where no matter what dog (in this example,
    # it will always be a mammal species
    species = 'Mammal'

    def __init__(self, breed, name, age):
        # attributes
        # take in the argument
        # assign it using self.attribute_name <- you call on this 
        self.breed = breed
        self.name = name

        # expects an int
        self.age = age

    # methods are functions defined inside the body of the class and they're used to perform operations that sometimes
    # utilise the actual attribute of the object we created.

    #operation/action --> method
    def bark(self):
        print("Woof!")
```

```python
my_dog = Dog(breed = 'Great Dane', name = 'Spark', age = 4)
```

```python
type(my_dog)
```

```python
my_dog.breed
```

```python
my_dog.name
```

```python
my_dog.age
```

```python
# class example 2 copied from above

class Dog():

    # class object attribute, this is where no matter what dog (in this example,
    # it will always be a mammal species
    species = 'Mammal'

    def __init__(self, breed, name, service):
        # attributes
        # take in the argument
        # assign it using self.attribute_name <- you call on this 
        self.breed = breed
        self.name = name

        # expects an int
        self.service = service

    # methods are functions defined inside the body of the class and they're used to perform operations that sometimes
    # utilise the actual attribute of the object we created.

    #operation/action --> method
    def bark(self, servicenumber):
        print("Woof! My name is {} and I am {}! My service number is {}! WOOF!".format(self.name, self.breed, servicenumber))
```

```python
service_dog = Dog('German Shepherd','Tank',5)
```

```python
type(service_dog)
```

```python
service_dog.species
```

```python
service_dog.name
```
