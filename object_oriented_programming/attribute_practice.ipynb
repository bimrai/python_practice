```python
class Circle():

    # any rule related to circle always true
    pi = 3.14

    def __init__(self, radius=1):
        self.radius = radius

    # m ethod
    def get_circumference(self):
        return self.radius * self.pi * 2
```

```python
my_circle = Circle()
```

```python
my_circle.pi
```

```python
my_circle.radius
```

```python
my_circle.get_circumference()
```

```python
# If you want to check different radius over the default value of 1:

circle_radius_40 = Circle(20)
```

```python
circle_radius_40.get_circumference()
```

```python
# You can override the previous existing variable instantiations value. So say we had my_circle
# we can override and add a new value.

my_circle = Circle(2) # it was 1 by default, we checked method output 6.28, now it will be different from that.
```

```python
my_circle.get_circumference()
```

```python
# If you have an attribute, it does't always neccessarily have to be defined from a particular parameter call

class Circle():

    # any rule related to circle always true
    pi = 3.14

    def __init__(self, radius=1):
        self.radius = radius
        self.area = radius*radius*Circle.pi # the example, if its a class object attribute you can use class name, otherwise use self

    # m ethod
    def get_circumference(self):
        return self.radius * self.pi * 2
```

```python
# So lets check what we are working with
circle_a = Circle(30)
```

```python
circle_a.pi
```

```python
circle_a.radius
```

```python
circle_a.get_circumference()
```

```python
circle_a.area # Notice we dont use parentheses because it is a self.area initialisation. Not a method.
```
