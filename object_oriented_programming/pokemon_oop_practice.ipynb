# Pokemon Class Practice

```python
class Pikachu():

    def __init__(self, name, element):
        self.name = name
        self.element = element

    def speak(self):
        return self.name + ": Pika Pika!"

    def get_element(self): 
        return "{} Type: {}".format(self.name, self.element)
```

```python
pikachu = Pikachu('Pikachu', 'Electric')
```

```python
pikachu.speak()
```

```python
pikachu.get_element()
```
